```python
from pathlib import Path
import json
import re
from tqdm import tqdm

settings = json.loads(Path('../../../settings.json').read_text())
preprocessed_data_path = Path(settings['preprocessed_data_path'])
data_path = Path('.').resolve()
data_name = data_path.name
source_name = data_path.parent.name
preprocessed_dir = preprocessed_data_path/source_name/data_name

source_data_dir = preprocessed_dir/'preprocessed'
splits = ['train', 'valid']
tasks = ['답변 근거 생성', 'LM(한국어)', '레퍼런스 기반 QA', 'QA 답변 교정']
task_data_dir = preprocessed_dir/'preprocessed_task'
task_data_dir.mkdir(exist_ok=True)
for task in tasks:
    task_path = task_data_dir/task
    task_path.mkdir(exist_ok=True)
    
#### prepare for task preprocess

pass

#### prepare for task preprocess end

#### task preprocess
for split in splits:
    source_data_dir_split = source_data_dir/split
    task_files = [(task_data_dir/task/f'{split}.jsonl').open('w', encoding='utf-8') for task in tasks]
    for source_data in tqdm(list(source_data_dir_split.iterdir()), desc=split):
        source_data = source_data.open()
        for line in source_data.readlines():
            line = json.loads(line)
            
            #### data preprocess
            for info in line['dataset']['context_info']:          
                for qa in info['qas']:
                    data = {'input': {}, 'output': {}}
                    data['input']['doc_id'] = line['dataset']['doc_id']
                    data['input']['context_id'] = info['context_id']
                    data['input']['context'] = info['context']
                    data['input']['question_id'] = qa['question_id']
                    data['input']['question-1'] = qa['question-1']
                    if (qa['question-2'] != None):
                        data['input']['question-2'] = qa['question-2']
            #### data preprocess end 

                    #### QA 답변 교정
                    # data = {'input': None, 'output': None}
                    ## preprocess data from line
                    data['output']['answer'] = []
                    data['output']['answer'].append(qa['answer'])
                    if 'wrong_answer' in qa:
                        for wrong in qa['wrong_answer']:
                            data['output']['answer'].append(wrong)
                    ## preprocess data from line end
                    task_files[3].write(json.dumps(data, ensure_ascii=False)+'\n')
                    # print(json.dumps(data, indent=4, ensure_ascii=False))
                    #### QA 답변 교정 end

                    #### 답변 근거 생성
                    # data = {'input': None, 'output': None}
                    ## preprocess data from line
                    data['input']['answer'] = qa['answer']
                    if 'answer_start' in qa:
                        data['input']['answer_start'] = qa['answer_start']
                        data['input']['answer_end'] = qa['answer_end']

                    del data['output']['answer']
                    data['output']['clue'] = qa['clue']

                    ## preprocess data from line end
                    task_files[0].write(json.dumps(data, ensure_ascii=False)+'\n')
                    # print(json.dumps(data, indent=4, ensure_ascii=False))
                    #### 답변 근거 생성 end

                    #### 레퍼런스 기반 QA
                    # data = {'input': None, 'output': None}
                    ## preprocess data from line
                    del data['input']['answer']
                    if 'answer_start' in qa:
                        del data['input']['answer_start']
                        del data['input']['answer_end']
                    del data['output']['clue']

                    data['input']['clue'] = qa['clue']
                    data['output']['answer'] = qa['answer']
                    if 'answer_start' in qa:
                        data['output']['answer_start'] = qa['answer_start']
                        data['output']['answer_end'] = qa['answer_end']

                    ## preprocess data from line end
                    task_files[2].write(json.dumps(data, ensure_ascii=False)+'\n')
                    # print(json.dumps(data, indent=4, ensure_ascii=False))
                    #### 레퍼런스 기반 QA end

            #### LM(한국어)
            data = {'text': None}
            ## preprocess data from line
            text = ''
            for info in line['dataset']['context_info']:     
                text += '##context:\n' + info['context'] + '\n'     
                for qa in info['qas']:
                    text += '##question-1:\n' + qa['question-1'] + '\n'
                    if (qa['question-2'] != None):
                        text += '##question-2:\n' + qa['question-2'] + '\n'
                    text += '##answer:\n' + qa['answer'] + '\n'
            data['text'] = text + '\n'
            ## preprocess data from line end
            task_files[1].write(json.dumps(data, ensure_ascii=False)+'\n')
            # print(json.dumps(data, indent=4, ensure_ascii=False))
            #### LM(한국어) end


            
    #         break
    #     break
    # break

    for path in task_files:
        path.close()      
        
#### task preprocess end      
```
